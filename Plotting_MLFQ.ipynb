```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 12]

file = open("temp", 'r')

content = file.readlines()

master = []
for i in range(8):
	master.append([])

for line in content:
	words = line.split(' ')
	if words[0] != '\n':
		master[int(words[0])-33].append(words[12])

x = []
for i in range(len(master[0])):
	x.append(50*i)

for i in range(len(master)):
    for j in range(len(master[i])):
        master[i][j] = int(master[i][j])
        
fig, ax = plt.subplots()
plt.xlabel("Time Elapsed (ticks)")
plt.ylabel("Queue number")
proc1, = ax.plot(x, master[0])
proc2, = ax.plot(x, master[1])
proc3, = ax.plot(x, master[2])
proc4, = ax.plot(x, master[4])
proc5, = ax.plot(x, master[5])

proc1.set_label("Process 1")
proc2.set_label("Process 2")
proc3.set_label("Process 3")
proc4.set_label("Process 4")
proc5.set_label("Process 5")

ax.legend()

plt.savefig("Plot.png")
```
